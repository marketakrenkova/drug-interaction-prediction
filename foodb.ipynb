### Cleaning data

```python
food.drop(
    ['description', 'picture_file_name', 'picture_content_type', 
    'picture_file_size', 'picture_updated_at', 'created_at', 'updated_at', 
    'creator_id', 'updater_id', 'export_to_afcdb', 'export_to_foodb'],
     axis=1, inplace=True)  
# food.head()
```

```python
food['itis_id'] = food['itis_id'].fillna(0).astype(int)  
food['legacy_id'] = food['legacy_id'].fillna(0).astype(int)
food['ncbi_taxonomy_id'] = food['ncbi_taxonomy_id'].fillna(0).astype(int)

food.head()  
```

```python
acc_num.drop(['created_at', 'updated_at', 'source_id', 'source_type'], axis=1, inplace=True)
acc_num.head()
```

```python
compound.drop(
    ['moldb_iupac', 'state', 'annotation_quality', 'cas_number', 
    'moldb_inchikey', 'moldb_mono_mass', 'moldb_smiles', 'moldb_inchi', 
    'kingdom', 'superklass', 'klass', 'subklass'], 
    axis=1, inplace=True)
# keep smiles?

compound.rename(columns={'description': 'cas_number'}, inplace=True)
compound.head()
```

```python
compound_ex_descriptor.drop(['creator_id', 'updater_id', 'created_at', 'updated_at', 'annotations'], axis=1, inplace=True)
compound_ex_descriptor.head()
```

```python
content.drop(
    ['citation', 'citation_type','creator_id', 'updater_id', 'created_at', 'updated_at',
    'preparation_type', 'export', 'orig_citation', 'orig_method', 'orig_unit_expression'], 
    axis=1, inplace=True)
content.head()
```

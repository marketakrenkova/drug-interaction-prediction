### Herb-drug Interactions Extraction
Extracting herb-drug interactions from https://www.mskcc.org/cancer-care/diagnosis-treatment/symptom-management/integrative-medicine/herbs/search.

The __herb-drug_interactions.json__ is created from a webscraped file __herbs_scrapping.json__. 

In this notebook, we extract names of drugs and map them to their ATC codes (using drug_id_name_map.csv, drug_synonyms.csv, drug_atc_code.csv). The output is a .tsv file in a following format: __\<drug's atc_code\>, interacts_with, \<herb\>__  

```python
import pandas as pd
import json
import re
import itertools
```

```python
with open('herb-drug_interactions.json', 'r') as f:
    herbs_drug_interactions = json.load(f)

herbs_drug_interactions
```

```python
herbs_drugs_df = pd.DataFrame({'herb': herbs_drug_interactions.keys(),
                               'drugs': [elem['drugs'] for elem in list(herbs_drug_interactions.values())]})
herbs_drugs_df.head()
```

```python
def transform_drug_names(drugs):
    new_drugs = [re.sub('-', ' ', drug).lower() for drug in drugs]
    new_drugs = list(set(new_drugs))
    return new_drugs
```

```python
herbs_drugs_df['drugs_transformed'] = herbs_drugs_df['drugs'].apply(transform_drug_names)
herbs_drugs_df.head()
```

```python
herbs_drugs_df = herbs_drugs_df.explode('drugs_transformed').dropna().drop(columns=['drugs'])
```

```python
# split by '(', ',', '/'
# frist, try match the first thing, then the second ...

splitted_drugs = []

for drug in list(herbs_drugs_df.drugs_transformed):
    parts = re.split(r',|\(|/|\)', drug)    
    splitted_drugs.append((parts))


splitted_drugs = [[d.strip() for d in drugs if d != ''] for drugs in splitted_drugs]

for split_d in splitted_drugs:
    for i, d in enumerate(split_d):
        if 'anticoagulant' in d:
             split_d[i] = 'anticoagulants'
        elif 'antiplatelet' in d:
            split_d[i] = 'antiplatelet agents'
        elif 'cyp450' in d or 'cyp 450' in d or 'cytochrome p450' in d:
            split_d[i] = 'cytochrome p-450'
            
# splitted_drugs
```

```python
herbs_drugs_df['splitted_drugs'] = splitted_drugs
herbs_drugs_df_transformed = herbs_drugs_df.explode("splitted_drugs")
herbs_drugs_df_transformed
```

```python
drug_names = pd.read_csv('../data/drugbank/drug_id_name_map.csv', index_col=[0])
print(drug_names.head())

drug_atc = pd.read_csv('../data/drugbank/drug_atc_code.csv', index_col=[0])
print(drug_atc.head())
```

```python
drug_synonyms = pd.read_csv('../data/drugbank/drug_synonyms.csv', index_col = [0])
drug_synonyms
```

```python
drug_category = pd.read_csv('../data/drugbank/drug_category.csv', index_col=[0])
drug_category['category'] = drug_category['category'].str.lower()
set(drug_category.category)
```

```python
drug_category[drug_category['category'] == 'diuretics']
```

```python
herbs_drugs_set = set(herbs_drugs_df_transformed.splitted_drugs)
drug_category_set = set(drug_category.category)
drug_category_set.intersection(herbs_drugs_set)
```

```python
drugs_dict = dict(zip(drug_names.drug_name, drug_names.id))
drugs_dict_lower = {}
for k, v in drugs_dict.items():
    new_k = re.sub('-', ' ', k)
    new_k = new_k.lower()
    drugs_dict_lower[new_k] = v

drugs_synonyms_dict = dict(zip(drug_synonyms.synonym, drug_synonyms.drug_id))
drugs_synonyms_dict_lower = {}
for k, v in drugs_synonyms_dict.items():
    new_k = re.sub('-', ' ', k)
    new_k = new_k.lower()
    drugs_synonyms_dict_lower[new_k] = v
```

```python
herbs_drugs_df_transformed['drugbank_id'] = list(itertools.repeat(None, herbs_drugs_df_transformed.shape[0]))
herbs_drugs_df_transformed['drug_category'] = list(itertools.repeat(None, herbs_drugs_df_transformed.shape[0]))
herbs_drugs_df_transformed = herbs_drugs_df_transformed.reset_index(drop=True)
herbs_drugs_df_transformed.head()
```

```python
# for extracted_drug in splitted_drugs:
for row in herbs_drugs_df_transformed.itertuples():
    d = row[3]
    # for d in extracted_drug:
    if d in drugs_dict_lower.keys():
        herbs_drugs_df_transformed['drugbank_id'][row[0]] = drugs_dict_lower[d]

    elif d in drugs_synonyms_dict_lower.keys():
        herbs_drugs_df_transformed['drugbank_id'][row[0]] = drugs_synonyms_dict_lower[d]
        
    elif d in drug_category_set:
        herbs_drugs_df_transformed['drug_category'][row[0]] = d
    elif 'cytochrome p-450' in d:
        herbs_drugs_df_transformed['drug_category'][row[0]] = d
```

```python
herbs_drugs_df_transformed
```

```python
herbs_drugs_df_transformed_atc = herbs_drugs_df_transformed.merge(drug_atc, left_on="drugbank_id", right_on="id", how="left").drop(columns=["id", "drugs_transformed", "drug_category"]).dropna()
herbs_drugs_df_transformed_atc
```

```python
herbs_drugs_atc = herbs_drugs_df_transformed_atc[["herb", "atc_code"]]
herbs_drugs_atc["relation"] = list(itertools.repeat("interacts_with", herbs_drugs_atc.shape[0]))
herbs_drugs_atc = herbs_drugs_atc.iloc[:,[1,2,0]]
herbs_drugs_atc = herbs_drugs_atc.rename(columns={"atc_code": "drug"})
herbs_drugs_atc = herbs_drugs_atc.reset_index(drop=True)
herbs_drugs_atc
```

```python
herbs_drugs_atc.to_csv("../data/triplets/herbs-di-atc.tsv", sep="\t")
```

```python
def choose_value(row):
    if row['drugbank_id'] is not None:
        return row['drugbank_id']
    else:
        return row['drug_category']

herbs_drugs_df_for_drugbank = herbs_drugs_df_transformed[["herb", "drugbank_id", "drug_category"]]
herbs_drugs_df_for_drugbank['drug_or_category'] = herbs_drugs_df_for_drugbank.apply(choose_value, axis=1)
herbs_drugs_df_for_drugbank = herbs_drugs_df_for_drugbank.drop(columns=["drugbank_id", "drug_category"]).dropna()
herbs_drugs_df_for_drugbank 
```

```python
# herbs_drugs_drugbank = herbs_drugs_df_transformed_atc[["herb", "drugbank_id"]]
herbs_drugs_df_for_drugbank["relation"] = list(itertools.repeat("interacts", herbs_drugs_df_for_drugbank.shape[0]))
herbs_drugs_df_for_drugbank = herbs_drugs_df_for_drugbank.iloc[:,[1,2,0]]
herbs_drugs_df_for_drugbank = herbs_drugs_df_for_drugbank.reset_index(drop=True)
herbs_drugs_df_for_drugbank = herbs_drugs_df_for_drugbank.rename(columns={'drug_or_category': 'drug1', 'relation': 'interaction', 'herb': 'drug2'})
herbs_drugs_df_for_drugbank.head()
```

```python
# add some interactions manually (important herbs for docters, that were not extracted automatically)
more_interactions = [
        ('anticoagulants', 'interacts', 'Cordyceps'),
        ('antiplatelet agents', 'interacts', 'Cordyceps'),
        ('insulin', 'interacts', 'Cordyceps'),
        ('oral hypoglycemics', 'interacts', 'Cordyceps'),
] 
new_data = pd.DataFrame(more_interactions, columns=['drug1', 'interaction', 'drug2'])
new_data
herbs_drugs_df_for_drugbank = pd.concat([herbs_drugs_df_for_drugbank, new_data]).reset_index(drop=True)
herbs_drugs_df_for_drugbank
```

```python
herbs_drugs_df_for_drugbank.to_csv("../data/triplets/herbs-di.tsv", sep="\t")
```

### Food contribution
For each food, get its compounds - order them according to contribution score (top 10??).

contribution score = compund content in the food / content of all compounds in the food (same as in https://www.nature.com/articles/s41598-022-05132-y.pdf)

UPDATE 1: for each food, use only list of compounds which are in phytochemicals.txt (these compounds are likely to inmteract with drugs)

UPDATE 2: use phytocheicals and compounds which are in drugs

```python
# UPDATE 1
with open('../data/phytochemicals.txt', 'r') as f:
    phytochemicals = f.readlines()
    
phytochemicals = [re.sub('\n', '', x) for x in phytochemicals]   
print(phytochemicals[:10])
```

```python
# UPDATE 2
drug_ingredients = pd.read_csv('../data/drugbank/drug_id_name_map.csv', index_col=[0])
drug_compounds = list(set(drug_ingredients['drug_name']))
drug_compounds = [x.lower().strip() for x in drug_compounds]
 
print(drug_compounds[:10])
 
phytochemicals += drug_compounds
```

```python
phytochemicals = list(set(phytochemicals))
```

```python
def get_phytochemicals(food_content, phytochemicals):
    compounds = []
    for compound_id in list(food_content.source_id):
        compound_name = food_content[food_content['source_id'] == compound_id].orig_source_name.values[0]
        if type(compound_name) is float:
            continue

        if compound_name.lower() in phytochemicals and compound_id not in compounds:
            compounds.append(compound_id)
    return compounds        
```

```python
# returns dictionary of top k most contributing compounds of the given food (food_content)
# compound_id : contribution score
def contribution_scores(food_content, k=10):
    compounds = dict()

    content_all = food_content[food_content['orig_unit'] == 'mg/100g'].orig_content.sum()
    
    for compound_id in list(food_content.source_id):
        score = 0 # no content information -> score = 0
        if content_all:
            score = food_content[food_content['source_id'] == compound_id].orig_content.values[0] / content_all
        compounds[compound_id] = score

    compounds = sorted(compounds.items(), key=lambda item: item[1], reverse=True)  
    
    top_compounds = dict(compounds[:k])
    return top_compounds
```

```python
# for each most contributing compound of a food returns:
# public_id, source_id, name of the compound, cas number and contribution score
def map_to_compounds(compounds, compound_df):
    compounds_list = list()
    
    for k in compounds:
        compound_info = compound_df[compound_df['id'] == k]
        if not compound_info.empty:
            compounds_list.append({
                'public_id': compound_info.public_id.values[0], 
                'source_id': k,
                'name': compound_info.name.values[0],
                'cas_number': compound_info.cas_number.values[0],
            })
            
    return compounds_list
```

```python
food_content = content[content['food_id'] == 4]
food_content = food_content.dropna(subset=['orig_content'])

# most_contribution_compounds = contribution_scores(food_content)
# kiwi_compounds = map_to_compounds(most_contribution_compounds)

compounds = get_phytochemicals(food_content, phytochemicals)
kiwi_compounds = map_to_compounds(compounds, compound_df)


print(kiwi_compounds)
```

```python
# returns most contributing compounds for all food
def food_compounds(phytochemicals, compound_df):
    food_ids = list(zip(food['public_id'].values, food['id'].values))
    food_compounds_dict = dict()

    for f_public_id, f_id in food_ids:
        food_content = content[content['food_id'] == f_id]
        most_contribution_compounds = get_phytochemicals(food_content, phytochemicals)
        food_compounds_dict[f_public_id] = map_to_compounds(most_contribution_compounds, compound_df)

    return food_compounds_dict    
```

```python
food_compounds_dict = food_compounds(phytochemicals, compound_df)
```
